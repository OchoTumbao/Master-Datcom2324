```r
temp <- Boston
plotY <- function (x,y) {
plot(temp[,y]~temp[,x], xlab=paste(names(temp)[x]," X",x,sep=""),
ylab=names(temp)[y])
}
par(mfrow=c(2,2)) #Si margin too large => (2,3)
x <- sapply(1:(dim(temp)[2]-1), plotY, dim(temp)[2])
par(mfrow=c(1,1))
```

```r
fit1=lm(medv~lstat,data=Boston)
fit1
```

```r
summary(fit1)
par(mfrow=c(2,1))
plot(medv~lstat,Boston)
abline(fit1,col="red")
confint(fit1)
```

```r
fit2=lm(medv~rm,data=Boston)
fit2
summary(fit2)
plot(medv~rm,Boston)
abline(fit2,col="blue")
par(mfrow=c(1,1))
confint(fit2)
```

```r
temp <- Boston
plot(temp[,-dim(temp)[2]],pch=16,col=gray(1-(temp[,dim(temp)[2]]/max(temp[,dim(temp)[2]]))))
```

```r
fit5=lm(medv~.,data=Boston)
print(summary(fit5))
```

```r
fit6=lm(medv~.-age,data=Boston)
print(summary(fit6))
```

```r
fit7=lm(medv~.-age-indus,data=Boston)
print(summary(fit7))
```

```r
fitprueba=lm(medv~lstat +rm +I(lstat * rm) +I(lstat^2) +I(lstat^2 * rm),Boston)
summary(fitprueba)
```

```r
plot(medv~lstat,Boston)
points(Boston$lstat,fitted(fitprueba),col="red",pch=20)
```

```r
fit8=lm(medv~lstat*rm,Boston)
summary(fit8)
plot(medv~lstat,Boston)
points(Boston$lstat,fitted(fit8),col="green",pch=20)
```
