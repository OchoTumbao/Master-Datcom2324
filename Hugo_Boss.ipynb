```r
library(tidyverse)
library(readxl)
```

```r
data <- read_excel("Hugo_boss.xlsx")
colnames(data)[10] <- "units_sold"
colnames(data)[2] <- "Line_name"
colnames(data)[11] <- "Sold_value"
colnames(data)
```

```r
?arrange
```

```r
data %>% group_by(Style,Color,Line_name) %>% summarise(Revenue_generated=sum(units_sold*Sold_value)) %>% arrange(desc(Revenue_generated)) %>% filter(Line_name=="Green")
```

```r
data %>% group_by(Style,Color,Line_name) %>% summarise(Revenue_generated=sum(units_sold*Sold_value)) %>% top_n(1,Revenue_generated)
```
