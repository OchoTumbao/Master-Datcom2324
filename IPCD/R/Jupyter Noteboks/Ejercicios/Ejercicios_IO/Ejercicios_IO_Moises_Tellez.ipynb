5. Cargue la primer y tercera hojas del fichero `resultados.xls` y muestre un gráfico que compare, para los dos datasets, el resultado en entrenamiento y test a medida que aumenta la cantidad de bits utilizados.

```r
data1 <-read_excel("results.xlsx", sheet = "MNIST Non-convolutional",col_names = TRUE, skip = 1)
data2 <-read_excel("results.xlsx", sheet = "Fashion MNIST Non-convolutional",col_names = TRUE, skip = 1)
data_plot <- full_join(data1,data2)
colnames(data_plot)[1:3] <- c("Dataset","Fold","Bits")
data_plot %>% pivot_longer(cols=c("Train","Test")) %>% ggplot(mapping=aes(x=Bits,y=value,group=name,color=Dataset)) + geom_point()
```
